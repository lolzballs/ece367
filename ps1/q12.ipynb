```julia
using LinearAlgebra
using Convex
using SCS
```

```julia
function proj(x, v0, v, norm_fn)
    solver = () -> SCS.Optimizer(verbose=0)
    y = Variable(2)
    t = Variable()
    p = minimize(norm_fn(x - y), y == t * v + v0)
    solve!(p, solver)
    return y.value, p.optval
end

v0 = [-2; -4]
v = [-2; 5]
x = [3; 2];
```

```julia
proj(x, v0, v, norm2)[1]
```

```julia
proj(x, v0, v, x -> norm(x, 1))[1]
```

```julia
proj(x, v0, v, x -> maximum(x))[1]
```

```python
import numpy as np
import scipy.io
```

Load matlab data

```python
data = scipy.io.loadmat('../ps1/data/wordVecV.mat')['V']
```

Generate $M$ and $\tilde{M}$

```python
M = (data > 0).astype(int)
M_norm = M / M.sum(0)
```

```python
U, S, V = np.linalg.svd(M_norm)
print(S[:10])
```
